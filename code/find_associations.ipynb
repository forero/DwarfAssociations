```python
%pylab inline
from scipy.cluster.vq import kmeans,vq, whiten
rcParams['figure.figsize'] = 10, 10
```

```python
data_path = "../data/B64_WM5_10909_LG_7Mpc_2048/"
catalog_name = data_path+"snap_134_z0.000.old_AHF_halos"
AHF_data = loadtxt(catalog_name)
shape(AHF_data)
```

```python
def rms_3D(vec_x, vec_y, vec_z):
    std_x = std(vec_x)
    std_y = std(vec_y)
    std_z = std(vec_z)
    return sqrt(std_x**2 + std_y**2 + std_z**2)
```

```python
#select the data in the high resolution region
n_part_1 = AHF_data[:,0]
n_part_2 = AHF_data[:,1]
index = where(n_part_1==n_part_2)
index = index[0]
AHF_data = AHF_data[index,:]
```

```python
x_pos = AHF_data[:,2]
y_pos = AHF_data[:,3]
z_pos = AHF_data[:,4]
vel_x = AHF_data[:,5]
vel_y = AHF_data[:,6]
vel_z = AHF_data[:,7]
mass = AHF_data[:,8]
```

```python
scatter(x_pos, y_pos, s=0.1)
```

```python
scatter(x_pos, z_pos, s=0.1)
```

```python
n_points =size(x_pos)
features = empty((n_points,3))
features[:,0] = x_pos[:]
features[:,1] = y_pos[:]
features[:,2] = z_pos[:]
```

```python
n_clusters = 200
centroids,_ = kmeans(features,n_clusters)
idx,_ = vq(features,centroids)
```

```python
scatter(x_pos, y_pos, s=0.1)
scatter(centroids[:,0],centroids[:,1], color='red')
```

```python
scatter(x_pos, z_pos, s=0.1)
scatter(centroids[:,0],centroids[:,2], color='red')
```

```python
sizes = empty((0))
rms_vel = empty((0))
rms_pos = empty((0))
total_mass = empty((0))
group_id = arange(n_clusters)
for i in range(n_clusters):
    id_group = where(idx==i)
    rms_vel = append(rms_vel, rms_3D(vel_x[id_group], vel_y[id_group], vel_z[id_group]))
    rms_pos = append(rms_pos, rms_3D(x_pos[id_group], y_pos[id_group], z_pos[id_group]))
    total_mass = append(total_mass, log10(sum(mass[id_group])))
    sizes = append(sizes, size(where(idx==i)))
  #  print size(where(idx==i)), i
#plot(sizes)
```

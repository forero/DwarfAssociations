```python
data = np.loadtxt("rmax.dat")
```

```python
x = data[:,0]
y = data[:,1]
z = data[:,2]
r = data[:,3]
```

```python
print x[0]
```

```python
theta = np.linspace(0, 2*np.pi, 100)
phi = np.linspace(0, np.pi, 100)
```

```python
for i in range(29):
    X = r[i+29]*cos(theta) + x[i+29]
    Y = r[i+29]*sin(theta) + y[i+29] 
    scatter(x[i+29], y[i+29])
    plot(X, Y)
    plt.xlabel("$X$", fontsize=20)
    plt.ylabel("$Y$", fontsize=20)
```

```python
for i in range(29):
    X = r[i]*cos(theta) + x[i]
    Z = r[i]*sin(theta) + z[i] 
    scatter(x[i], z[i])
    plot(X, Z)
    plt.xlabel("$X$", fontsize=20)
    plt.ylabel("$Z$", fontsize=20)
```

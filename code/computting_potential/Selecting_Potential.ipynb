```python
%pylab inline
from scipy.interpolate import griddata
```

```python
data = np.loadtxt("../../data/B64_WM5_10909_LG_7Mpc_2048/potenciales.txt")
data2 = np.loadtxt("../../data/B64_WM5_10909_LG_7Mpc_2048/massandradius.txt")
```

```python
x = data[:,0]
y = data[:,1]
z = data[:,2]
u = data[:,3]
M = data2[:,0]
R = data2[:,1]
```

```python
scatter(x, y)
```

```python
index2 = np.where((x>11) & (x<30) & (y>22) & (y<45) & (z>23) & (z<45))

x = x[index2]
y = y[index2]
z = z[index2]
u = u[index2]
M = M[index2]
R  =R[index2]
```

```python
xmin = amin(x)
xmax = amax(x)
ymin = amin(y)
ymax = amax(y)
```

```python
figure(num=None, figsize=(9.5,8))
points = (x, y)
grid_x, grid_y = mgrid[xmin:xmax:200j, ymin:ymax:200j]
grid_zA = griddata(points, log10(u), (grid_x, grid_y), method='cubic')
imshow(grid_zA.T, extent=(xmin,xmax,ymin,ymax), origin='lower')
colorbar()
plt.xlabel("$x(Mpc)$",fontsize=35)
plt.ylabel("$y(Mpc)$",fontsize=35)
#scatter(x, y, s=5)
```

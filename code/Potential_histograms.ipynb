```python
data = np.loadtxt("../data/Distances/potenciales.txt")
```

```python
x = data[:, 0]
y = data[:, 1]
z = data[:, 2]
u = data[:, 3]
```

```python
plt.scatter(x, y, s=0.1)
```

```python
ymin = min(y)
xmin = min(x)
xmax = max(x)
ymax = max(y)
```

```python
fig = figure(num=None, figsize=(19, 19.5))
#plt.scatter(x, y, s=1)
aspectratio = 1.0*(max(x) - min(x))/(1.0*max(y) - min(y))
hist,xedges,yedges = numpy.histogram2d(x,y,bins=(400,400), weights=u,range=[[ymin,ymax],[xmin,xmax]])
hist = hist.transpose()
my_extent = (xedges[0], xedges[-1], yedges[0], yedges[-1])
imshow(log(hist),extent=[xmin, xmax, ymax, ymin], interpolation='nearest', aspect=aspectratio)
plt.xlabel("$x$", fontsize = 30)
plt.ylabel("$y$", fontsize = 30)
colorbar(shrink = 0.55)
plt.xlim(10, 35)
plt.ylim(24, 42)
plt.savefig("LGpotential.png")
```

```python
fig = figure(num=None, figsize=(19, 19.5))
cs = contour(log(hist), origin='lower',levels = np.linspace(55, 60, 5), extent=[xmin, xmax, ymax, ymin])
plt.clabel(cs, fontsize=19)
plt.xlim(10, 35)
plt.ylim(24, 42)
```
